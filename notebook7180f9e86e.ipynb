```python
training_data=[]
for examples in data['examples']:
    temp_dict={}
    temp_dict['text']=examples['content']
    temp_dict['entities']=[]
    for annotations in examples['annotations']:
        start=annotations['start']
        tag_name=annotations['tag_name'].upper()
        end=annotations['end']
        temp_dict['entities'].append((start,end,tag_name))
    training_data.append(temp_dict)
```

```python
we will first import DocBin from spacy.tokens in whose object
the final doc will be stored then an empty model
is loaded with nlp=spacy.blank("en")
tqdm is imported from tqdm,an object of DocBin() is created
filter_spans is imported from spacy.util
training_data is loaded in tqdm like tqdm(training_data)
and now for every element of tqdm(training_data)
text,entities will be extracted and stored in text and labels respectively
a doc element is made out of text using nlp.make_doc(text)
now all labels which is a list having (start,end,labels) each of
this is acessed and respective span object is made out of this
using doc.char_span(start,end,label=label,alignment_mode="contract")
if span is None then rejected else span is appened to ents list
now the duplicates are removed using filter_spans filter_spans(ents)
and doc.ents is initialized with filtered_ents, Lastly the doc
object which was initially made out of text using nlp.make_doc(text)
is lastly added to doc_bin as doc_bin.add(doc)
and after coming out of both loops 
doc_bin.to_disk("train.spacy")
```

```python
from spacy.tokens import DocBin
from tqdm import tqdm
nlp=spacy.blank('en')
doc_bin=DocBin()
from spacy.util import filter_spans

for training_example in tqdm(training_data):
    text=training_example['text']
    labels=training_example['entities']
    doc=nlp.make_doc(text)
    ents=[]
    for start,end,label in labels:
        span=doc.char_span(start,end,label=label,alignment_mode="contract")
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
    filtered_ents=filter_spans(ents)
    doc.ents=filtered_ents
    doc_bin.add(doc)
doc_bin.to_disk("train.spacy")
```
